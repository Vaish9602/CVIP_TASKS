**Normal Task: Analyze customer behavior based on transaction data

Develop a data analysis project that focuses on understanding and analyzing
customer behavior based on transactional data. The goal is to perform Exploratory

purchasing patterns, and overall behavior.**

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Hackathon_Ideal_Data.csv")
```

```python
df
```

```python

df.head()
```

```python
df.info()
```

```python
# Perform Exploratory Data Analysis (EDA)
# Example 1: Distribution of Purchase Amounts
plt.figure(figsize=(10, 6))
sns.histplot(df['VALUE'], bins=30, kde=True)
plt.title('Distribution of Purchase Amounts')
plt.xlabel('VALUE')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example 2: Monthly Total Purchase Amount
df['QTY'] = pd.to_datetime(df['QTY'])
df['month'] = df['QTY'].dt.to_period('M')

monthly_purchase = df.groupby('month')['VALUE'].sum()

plt.figure(figsize=(12, 6))
monthly_purchase.plot(marker='o')
plt.title('Monthly Total Purchase Amount')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example 3: Customer Purchase Patterns
customer_purchase = df.groupby('STORECODE')['VALUE'].sum()

plt.figure(figsize=(12, 6))
sns.boxplot(x=customer_purchase.values)
plt.title('Customer Purchase Patterns')
plt.xlabel('Total Purchase Amount')
plt.show()
```

```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
stock_data=pd.read_csv("/content/dataset_tk.csv")
```

```python
stock_data
```

```python
# Display basic information about the dataset
print(stock_data.info())
```

```python
# Display the first few rows of the dataset
print(stock_data.head())
```

```python
def get_stock_data(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data
```

```python
ticker = 'AAPL'
start_date = '2010-01-01'
end_date = '2023-12-31'
stock_data = get_stock_data(ticker, start_date, end_date)
```

```python
# Display basic information about the dataset
print(stock_data.info())

# Display the first few rows of the dataset
print(stock_data.head())
```

```python
# Perform Time Series Analysis
# Example 1: Plotting Stock Prices Over Time
plt.figure(figsize=(14, 6))
stock_data['Close'].plot(label='Closing Price')
plt.title(f'{ticker} Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.show()
```

```python
# Example 2: Decomposition of Time Series (Trend, Seasonal, Residual)
decomposition = sm.tsa.seasonal_decompose(stock_data['Close'], model='additive', period=252)  # Assuming daily data with 252 trading days
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
plt.figure(figsize=(14, 8))
```

```python
plt.subplot(411)
plt.plot(stock_data['Close'], label='Original')
plt.title('Original Time Series')
```

```python
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.title('Trend Component')
```

```python
plt.subplot(413)
plt.plot(seasonal, label='Seasonal')
plt.title('Seasonal Component')
```

```python
plt.subplot(414)
plt.plot(residual, label='Residual')
plt.title('Residual Component')
```

```python
plt.tight_layout()
plt.show()
```
