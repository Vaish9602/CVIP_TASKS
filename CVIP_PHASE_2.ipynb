Develop a robust Spam Email Filter using Natural Language Processing (NLP)
techniques and machine learning algorithms. The goal is to create an intelligent
system capable of accurately classifying emails as either spam or legitimate
(ham) based on their content and linguistic features.**

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv('/content/emails.csv')

df
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['text'], test_size=0.2, random_state=42)
```

```python
# Build a spam filter using a Naive Bayes classifier with a bag-of-words model
model = make_pipeline(CountVectorizer(), MultinomialNB())
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the performance of the spam filter
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy:.2f}')
```

```python
print('Confusion Matrix:\n', conf_matrix)
```

```python
print('Classification Report:\n', classification_rep)
```
